```python
import numpy as np
```

```python
def conv2D(matrix:np.array,kernel:np.array):
    # get the dimensions of the matrix and kernel:
    m_rows, m_cols = matrix.shape
    k_rows, k_cols = kernel.shape
    
    # no padding, so the output dimensions will be:
    out_rows = m_rows - k_rows + 1 + 2
    out_cols = m_cols - k_cols + 1 + 2
    
    # initialize the output matrix:
    output = np.zeros((out_rows, out_cols))
    
    for i in range(out_rows):
        for j in range(out_cols):
            # extract the region of interest:
            region = matrix[i:i+k_rows, j:j+k_cols]
            # perform element-wise multiplication and sum:
            output[i, j] = np.clip(np.round(np.sum(region * kernel)), 0, 255)
    
    return output
```

```python
original_matrix = np.array([
    [0, 10, 1, 8, 1, 8, 8, 1],
    [8, 0, 8, 1, 28, 0, 10, 1],
    [29, 10, 29, 28, 28, 29, 28, 1],
    [3, 10, 3, 1, 5, 0, 3, 6],
    [28, 0, 28, 29, 1, 1, 10, 15],
    [28, 1, 29, 10, 10, 1, 10, 1],
    [29, 28, 28, 0, 12, 29, 3, 21],
    [29, 30, 3, 1, 5, 0, 1, 6]
])
```

# box filter

```python
kernel_box = np.array([
    [1/9, 1/9, 1/9],
    [1/9, 1/9, 1/9],
    [1/9, 1/9, 1/9]
])
```

```python
# citra hasil menggunakan kernel box:
result_box = conv2D(original_matrix, kernel_box)
print("Hasil konvolusi dengan kernel box: (dibulatkan ke bilangan bulat terdekat)")
print(result_box)
```

# Weighted average Filter

```python
weighted_kernel = np.array([
    [1/16, 2/16, 1/16],
    [2/16, 4/16, 2/16],
    [1/16, 2/16, 1/16]
])
```

```python
# citra hasil menggunakan kernel weighted:
result_weighted = conv2D(original_matrix, weighted_kernel)
print("\nHasil konvolusi dengan kernel weighted: (dibulatkan ke bilangan bulat terdekat)")
print(result_weighted)
```

# Mask/Filter konvolusi

```python
mask_kernel = np.array([
    [1, 1, 1],
    [1, 3, 1],
    [1, 1, 1]
])
```

```python
result_mask = conv2D(original_matrix, mask_kernel)
print("\nHasil konvolusi dengan kernel mask: (dibulatkan ke bilangan bulat terdekat)")
print(result_mask)
```

# Citra menggunakan Max,min, median Filter (3x3)

```python
def filter(matrix:np.array,filter_type:str):
    M,N = matrix.shape
    
    window = np.zeros(9)
    image_result = np.zeros((M-2,N-2))
    
    for i in range(1,M-1):
        for j in range(1,N-1):
            window[0] = matrix[i-1][j-1]
            window[1] = matrix[i-1][j]
            window[2] = matrix[i-1][j+1]
            window[3] = matrix[i][j-1]
            window[4] = matrix[i][j]
            window[5] = matrix[i][j+1]
            window[6] = matrix[i+1][j-1]
            window[7] = matrix[i+1][j]
            window[8] = matrix[i+1][j+1]
            
            if filter_type == "median":
                image_result[i-1][j-1] = np.median(window)
            elif filter_type == "max":
                image_result[i-1][j-1] = np.max(window)
            elif filter_type == "min":
                image_result[i-1][j-1] = np.min(window)

    return image_result
    
```

```python
max_result = filter(original_matrix, "max")
print("\nHasil filter max:")
print(max_result)
```

```python
median_result = filter(original_matrix, "median")
print("\nHasil filter median:")
print(median_result)
```

```python
min_result = filter(original_matrix, "min")
print("\nHasil filter min:")
print(min_result)
```
